```python
import requests

url = "https://api.sportradar.com/ncaafb/trial/v7/en/polls/AP25/2025/01/rankings.json"

headers = {
    "accept": "application/json",
    "x-api-key": "NHtixgcMvSLeWvhTj8INJULDUrdevCsPHF6x8JYd"
}

response = requests.get(url, headers=headers)
```

```python
data = response.json()
data
```

```python
poll_id = data.get("poll",{}).get("id")
poll_name = data.get("poll",{}).get("name")
poll_alias = data.get("poll",{}).get("alias")
season = data.get("season")
week = data.get("week")
effective_date = data.get("effective_date")
```

```python
import pandas as pd
#extracting team rankings
rankings = data.get("rankings",[])

if rankings:
    rankings_info = []
    for r in rankings:
        ranking_details = {
            "poll_id": poll_id,
            "poll_name": poll_name,
            "poll_alias": poll_alias,
            "season": season,
            "week": week,
            "effective_date": effective_date,
            "team_id": r.get("id"),
            "team_name": r.get("name"),
            "market": r.get("market"),
            "rank": r.get("rank"),
            "points": r.get("points"),
            "fp_votes": r.get("fp_votes")
        }
        rankings_info.append(ranking_details)
    rankings_df = pd.DataFrame(rankings_info)
else:
    rankings_df = pd.DataFrame()
```

```python
import requests
import time

Base_url = "https://api.sportradar.com/ncaafb/trial/v7/en/teams/{team_id}/full_roster.json"

headers = {
    "accept": "application/json",
    "x-api-key": "NHtixgcMvSLeWvhTj8INJULDUrdevCsPHF6x8JYd"
}

# prepare lists for each entity
teams_list, venues_list, divisions_list, conferences_list, coaches_list, players_list = [], [], [], [], [], []

for i, row in teams_df.iterrows():
    team_id = row['team_id']
    url = Base_url.format(team_id=team_id)

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        data = response.json()

        # ========================
        # 1) TEAMS TABLE
        # ========================
        teams_list.append({
            "team_id": data.get("id"),
            "team_name": data.get("name"),
            "market": data.get("market"),
            "alias": data.get("alias"),
            "founded": data.get("founded"),
            "mascot": data.get("mascot"),
            "fight_song": data.get("fight_song"),
            "championships_won": data.get("championships_won"),
            "conference_titles": data.get("conference_titles"),
            "playoff_appearances": data.get("playoff_appearances"),
            "venue_id": data.get("venue", {}).get("id"),
            "division_id": data.get("division", {}).get("id"),
            "conference_id": data.get("conference", {}).get("id")
        })

        # ========================
        # 2) VENUES TABLE
        # ========================
        venue = data.get("venue", {})
        if venue:
            venues_list.append({
                "venue_id": venue.get("id"),
                "venue_name": venue.get("name"),
                "city": venue.get("city"),
                "state": venue.get("state"),
                "country": venue.get("country"),
                "zip": venue.get("zip"),
                "address": venue.get("address"),
                "capacity": venue.get("capacity"),
                "surface": venue.get("surface"),
                "roof_type": venue.get("roof_type"),
                "latitude": venue.get("location", {}).get("lat"),
                "longitude": venue.get("location", {}).get("lng")
            })

        # ========================
        # 3) DIVISIONS TABLE
        # ========================
        division = data.get("division", {})
        if division:
            divisions_list.append({
                "division_id": division.get("id"),
                "division_name": division.get("name"),
                "division_alias": division.get("alias")
            })

        # ========================
        # 4) CONFERENCES TABLE
        # ========================
        conference = data.get("conference", {})
        if conference:
            conferences_list.append({
                "conference_id": conference.get("id"),
                "conference_name": conference.get("name"),
                "conference_alias": conference.get("alias")
            })

        # ========================
        # 5) COACHES TABLE
        # ========================
        for c in data.get("coaches", []):
            coaches_list.append({
                "coach_id": c.get("id"),
                "team_id": team_id,
                "full_name": c.get("full_name"),
                "first_name": c.get("first_name"),
                "last_name": c.get("last_name"),
                "position": c.get("position")
            })

        # ========================
        # 6) PLAYERS TABLE
        # ========================
        for p in data.get("players", []):
            players_list.append({
                "player_id": p.get("id"),
                "team_id": team_id,
                "first_name": p.get("first_name"),
                "last_name": p.get("last_name"),
                "name": p.get("name"),
                "abbr_name": p.get("abbr_name"),
                "jersey": p.get("jersey"),
                "position": p.get("position"),
                "height": p.get("height"),
                "weight": p.get("weight"),
                "birth_place": p.get("birth_place"),
                "status": p.get("status"),
                "eligibility": p.get("eligibility")
            })

        print(f" {i+1}/{len(teams_df)}: {data.get('name')} parsed successfully")

    except Exception as e:
        print(f" Error fetching {team_id}: {e}")

    # Respect API rate limits
    time.sleep(1)
```
